### Merge all 10 runs

```python
allruns = []
ally = []
def integratedata(file_name):
    data = mne.io.read_raw_gdf('S01_MI/'+file_name, preload=True) 
    
    event_list = [1536, 1537, 1538, 1539, 1540, 1541, 1542]
    f_s = 512
    t_s = -2.0 #Trials of 5 seconds or Full length trial
    t_f = 3.0
    sample_s = int(t_s*f_s)
    sample_f = int(t_f*f_s)

    x = []
    y = []

    #have to run code below this for all 10 runs for subject1
    #data = mne.io.read_raw_gdf('S01_MI/motorimagination_subject1_run1.gdf') 
    eeg_data = data.get_data()
    # info = data.info

    event_info = data._annotations.description
    event_onset = data._annotations.onset
    time_values = data.times
    # print(eeg_data.shape)

    for e in range(len(event_info)):
      channel_data=[]
      if (int(event_info[e]) in event_list):
        event_sample = int(event_onset[e]*f_s)
        for ch in range(eeg_data.shape[0]):
          channel_data.append(eeg_data[ch, event_sample+sample_s:event_sample+sample_f])
        x.append(np.reshape(channel_data,(96,2560)))
        y.append(event_list.index(int(event_info[e])))
    allruns.append(x)
    ally.append(y)
```

```python
for i in range(10):
    file_name = f"motorimagination_subject1_run{i+1}.gdf"
    integratedata(file_name)
```

```python
print(np.array(allruns).shape)
x_new=[[0 for i in range(2560*42*10)] for j in range(64)]
for i in range(64):
    for j in range(10):
      for k in range(42):
        for l in range(2560):
          x_new[i][2560*42*j+2560*k+l] = allruns[j][k][i][l]
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Apr  2 12:46:07 2021

@author: INFINITE-WORKSTATION
"""
#import mne

#Just display errors and warnings
mne.set_log_level("WARNING")

#Loading EEG data
#data = mne.io.read_raw_gdf("motorimagination_subject1_run1.gdf", preload=True) 

eeg = x_new

s_freq = 512
eeg_chl = ['F3','F1','Fz','F2','F4','FFC5h','FFC3h','FFC1h','FFC2h','FFC4h',
           'FFC6h','FC5','FC3','FC1','FCz','FC2','FC4','FC6','FTT7h','FCC5h',
           'FCC3h','FCC1h','FCC2h','FCC4h','FCC6h','FTT8h','C5','C3','C1','Cz',
           'C2','C4','C6','TTP7h','CCP5h','CCP3h','CCP1h','CCP2h','CCP4h',
           'CCP6h','TTP8h','CP5','CP3','CP1','CPz','CP2','CP4','CP6','CPP5h',
           'CPP3h','CPP1h','CPP2h','CPP4h','CPP6h','P3','P1','Pz','P2','P4',
           'PPO1h','PPO2h','eog-l','eog-m','eog-r']

info = mne.create_info(eeg_chl, s_freq, ch_types=["eeg"] * 61 + ["eog"]*3)
raw = mne.io.RawArray(eeg, info)
raw.set_montage("standard_1005")
raw.set_eeg_reference('average', projection=True)

raw_tmp = raw.copy()
raw_tmp.filter(1,40, fir_design='firwin', skip_by_annotation='edge')
```

```python

#####################  Calling ICA ###################################

ica = mne.preprocessing.ICA(method="infomax",
                            fit_params={"extended": True},
                            random_state=1)#, n_components = 32)

#######################################################################
#################### ploting PSD ######################################

raw.plot_psd(area_mode='range', tmax=10.0, average=False)
raw_tmp.plot_psd(area_mode='range', tmax=10.0)

########################################################################
################### ICA Analysis #######################################

```

```python
ica.fit(raw_tmp)
ica.plot_sources(inst=raw_tmp)
ica.plot_components(inst=raw_tmp, picks=None) #to pick all channels - none
```

```python
ica.exclude = [0,4,5]

raw_corrected = raw_tmp.copy()
ica.apply(raw_corrected)

#########################################################################
################## Ploting filttred and Pre-prcossed  EEG ###############

raw.plot(n_channels=64, start=0.0, duration=10, scalings=dict(eeg=500e-6))
raw_corrected.plot(n_channels=64, start=0.0, duration=10,scalings=dict(eeg=500e-6))

eeg_preprocess = raw_corrected.get_data()

```

```python
x_filtered = np.load("filtered_X.npy")#raw_corrected._data
print(x_filtered.shape)
allruns_filtered = np.reshape(x_filtered,(64,10,42,2560))
x_train = [[[0 for k in range(2560)] for j in range(9)] for i in range(420)]
```

```python
print(np.array(x_train).shape)
channels = [0,2,4,27,29,31,54,56,58]
for i in range(9):    
    ch = channels[i]
    for j in range(10):
        for k in range(42):
            for l in range(2560):
                x_train[j*42+k][i][l]=allruns_filtered[ch][j][k][l]
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(9, 2560)),
    tf.keras.layers.Normalization(axis = None),
    
    #tf.keras.layers.Dropout(0.2, input_shape=(9*2560,)),
    tf.keras.layers.Dense(64, activation='relu',kernel_regularizer=tf.keras.regularizers.L2(0.1)),
    
    tf.keras.layers.Dense(7,activation = 'softmax'),
])
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=15)


model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
y_train = np.reshape(ally,(420,))
for i in range(10):
    for j in range(42):
        y_train[i*42+j] = ally[i][j]
#y_train = np.transpose(y_train)
x_train = np.array(x_train)
#model.fit(x_train, y_train, epochs=1000)   

from sklearn.model_selection import train_test_split

data_train, data_test, labels_train, labels_test = train_test_split(x_train, y_train, test_size=0.20, random_state=42)
model.fit(data_train, labels_train,validation_split = 0.2, epochs=1000)#,callbacks=[callback])   
```

```python

test_loss, test_acc = model.evaluate(data_test,labels_test, verbose=2)

print('\nTest accuracy:', test_acc)
```
